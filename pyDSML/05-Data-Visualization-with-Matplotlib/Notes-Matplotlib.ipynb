```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional
plt.plot(x,y) # 3rd argument for line and color stylization
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show() # not needed if on jupyter notbook with %matplotlib inline
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # Left Bottom Width Height
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Thic Plot')


axes2.plot(y,x)
axes2.set_title('Smol Plot')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2) # fig,axes <- tuple unpacking?

axes[0].plot(x,y)
axes[0].set_title('Furst Plot')

axes[1].plot(y,x)
axes[1].set_title('Other Plot')



#for current_ax in axes:
#    current_ax.plot(x,y)

#axes.plot(x,y)

#plt.tight_layout() <- takes care of overlapping plots
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend(loc='best') # equivilant to loc=0 [0-10] options on position # can do loc=(0.1,0.1) for specific place

#ax.set_title('Title')
#ax.set_ylabel('Ye')
#ax.set_xlabel('Xi')
```
